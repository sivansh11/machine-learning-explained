# Gradient Decent
Gradient decent outside the context of machine learning is a method of finding the local minima of a function 

```python
import matplotlib.pyplot as plt
```

lets take an example function
y = (x + 5)<sup>2</sup><br>
dy = 2 * (x + 5)

```python
def y(x: float) -> float:
    return (x + 5) ** 2

def dy(x: float) -> float:
    return 2 * (x + 5)
```

```python
# used to control the number of samples
samples = 100
```

```python
# graphing function y
plt.scatter([x for x in range(-100, 100, int(200 / samples))], [y(x) for x in range(-100, 100, int(200 / samples))])
```

Gradient decent is an optimisation algorithm, so lets take a random point x

```python
current_best_x = -75
```

```python
plt.scatter([x for x in range(-100, 100, int(200 / samples))], [y(x) for x in range(-100, 100, int(200 / samples))])
plt.scatter(current_best_x, y(current_best_x), color='yellow')
```

The local minima of a function is in the **-ve direction of the derivative** of that function or the **-ve direction of the gradient** of that function

so at some point x <br>
the local minima is in the -dy/dx direction

which means, that x - (dy/dx * some multiplier) is closer to the gradient than x

```python
new_x = current_best_x - (dy(current_best_x) * 0.1)   # 0.1 is some multiplier
plt.scatter([x for x in range(-100, 100, int(200 / samples))], [y(x) for x in range(-100, 100, int(200 / samples))])
plt.scatter(current_best_x, y(current_best_x), color='yellow')
plt.scatter(new_x, y(new_x), color='red')
```

we can repeat this process to get better and better values of x that gets closer and closer to the local minima (provided that the multiplier is small)

```python
all_new_x = []
x = current_best_x
for i in range(100):
    x = x - (dy(x) * 0.1)
    all_new_x.append(x)

plt.scatter([x for x in range(-100, 100, int(200 / samples))], [y(x) for x in range(-100, 100, int(200 / samples))])
plt.scatter(all_new_x, [y(x) for x in all_new_x], color='orange')
```

what if u use a very high multiplyer ?

```python
all_new_x = []
x = current_best_x
for i in range(100):
    x = x - (dy(x) * .9)
    all_new_x.append(x)

plt.scatter([x for x in range(-100, 100, int(200 / samples))], [y(x) for x in range(-100, 100, int(200 / samples))])
plt.scatter(all_new_x, [y(x) for x in all_new_x], color='orange')
```

as you can see, the new "optimised" points are all over the place, the value quite litrally bounces around the local minima

## Summary
Gradient decent is a optimization method to find the local minima of a function f(x) computationally

### step1:
first we start with a random guess for the local minima   

### step2:
we update our guess based on the fact that the local minima is in the -ve direction of the gradient of the function at x<br>
so <br>
new_x = x - (dy/dx * multiplyer)<br>
note: the multiplier used should be small, there is no fixed value, you can use what ever works for you

### step3:
we repeat this for n iterations
